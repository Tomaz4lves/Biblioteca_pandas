```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import chart_studio.plotly as py
import cufflinks as cf
import plotly.graph_objects as go
import plotly.express as px
from datetime import date
```

```python
commerce = pd.read_csv("data.csv", encoding="cp1252")
commerce.head()
```

```python
commerce.describe()
```

```python
nulo = commerce.isna().sum()
print(nulo)
```

```python
sn_commerce = commerce = commerce.dropna()
nulo = commerce.isna().sum()
print(nulo)
```

```python
commerce.describe()
```

```python
commerce.info()
```

```python
commerce = commerce[commerce['UnitPrice'] > 1]
```

```python
commerce = commerce[commerce['Quantity'] > 1]
```

```python
commerce.info()
```

```python
commerce = commerce.drop_duplicates()
```

```python
commerce.info()
```

```python
commerce['Quantity'] = commerce['Quantity'].astype('int64')
commerce.info()
```

```python
commerce['CustomerID'] = commerce['CustomerID'].astype('int64')
commerce.info()
```

```python
commerce['InvoiceDate'] = pd.to_datetime(commerce['InvoiceDate'])
commerce.info()
```

```python
commerce = commerce[commerce['Quantity'] < 10000]
commerce.info()
```

```python
commerce = commerce[commerce['UnitPrice'] < 5000]
commerce.info()
```

```python
commerce.describe()
```

```python
commerce['TotalPrice'] = commerce['Quantity'] * commerce['UnitPrice']
```

```python
commerce['TotalPrice'] = commerce['TotalPrice'].astype('int64')
commerce.info()
```

```python
final_date = commerce['InvoiceDate'].max()
print(final_date)
```

```python
num_valores_unicos = commerce['Country'].nunique()
print(num_valores_unicos)
```

```python
top10country = commerce.groupby('Country')['TotalPrice'].sum().nlargest(10)
print(top10country)
```

```python
fig = px.bar(top10country, x=top10country.index, y='TotalPrice')
fig.update_layout(xaxis_title='País', yaxis_title='Total de Vendas')
fig.show()
```

```python
top10product = commerce.groupby('Description')['Quantity'].sum().nlargest(10)
print(top10product)
```

```python
fig = px.bar(commerce.groupby('Description')['Quantity'].sum().nlargest(10))
fig.update_layout(xaxis_title='Produtos', yaxis_title='Quantidade de venda')
fig.show()
```

```python
commerce['month'] = commerce['InvoiceDate'].dt.month
SalesDate = commerce.groupby('month')['TotalPrice'].sum()
print(SalesDate)
```

```python
fig = px.bar(commerce.groupby('month')['TotalPrice'].sum())
fig.update_layout(xaxis_title='Meses', yaxis_title='Valor total de vendas')
fig.show()
```

```python
top_paises = commerce[commerce['Country'].isin(top10country.index)]
SCD = top_paises.groupby(['month', 'Country'])['TotalPrice'].sum().unstack()
print(SCD.head)
```

```python
fig = px.bar(top_paises.groupby(['month', 'Country'])['TotalPrice'].sum().unstack())
fig.update_layout(xaxis_title='Meses', yaxis_title='Valor total de vendas')
fig.update_xaxes(dtick=1)
fig.show()
```

```python
#Calculo RFM, confesso que tive dificuldades com os cálculos!

Recency = commerce.groupby(['CustomerID', 'InvoiceNo']).agg({'InvoiceDate': 'first', 'TotalPrice': 'sum'}).reset_index()
Recency['Recency'] = (final_date - Recency['InvoiceDate'])
Frequency = Recency.groupby('CustomerID')['InvoiceNo'].count().reset_index()
Frequency.columns = ['CustomerID', 'Frequency']
Monetary = Recency.groupby('CustomerID')['TotalPrice'].mean().reset_index()
Monetary.columns = ['CustomerID', 'Monetary']
RFM = Frequency.merge(Monetary, on='CustomerID').merge(Recency[['CustomerID', 'Recency']], on='CustomerID')
print(RFM)
```
