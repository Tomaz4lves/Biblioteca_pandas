```python
import pandas as pd
import numpy as np

# Carregar os dados do banco de dados
casa = pd.read_excel("house_2.xlsx")
casa.head()
```

```python
casa = casa.dropna()
```

```python
# Verificar e lidar com valores nulos
print('Valores nulos antes da limpeza:')
print(casa.isnull().sum())
```

```python
casa.describe()
```

```python
#Codificar os dados usando one-hot encoding
casa = pd.get_dummies(casa)
casa.head()
```

```python
#RÃ³tulos
labels = np.array(casa['SalePrice'])
#Removendo o rÃ³tulo da caracterÃ­sticas
features= casa.drop('SalePrice', axis = 1)
#Salvando o nome da categoria para uso posterior
feature_list = list(casa.columns)
#Converte para um array numpy
casa = np.array(features)
```

```python
#Importando libs e criando testes
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
from sklearn.datasets import make_moons
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from sklearn import metrics
```

```python
#GradientBoostingRegressor
grb = GradientBoostingRegressor(n_estimators=100)
grb.fit(train_features, train_labels)
gbr_pred = grb.predict(test_features)
```

```python
#Calculando erro absoluto
errors_gbr = abs(gbr_pred - test_labels)

r_sq = grb.score(test_features, test_labels)
print('Coeficiente de DeterminaÃ§Ã£o (RÂ²):', r_sq)

print('MAE:', metrics.mean_absolute_error(test_labels, gbr_pred))
print('MSE:', metrics.mean_squared_error(test_labels, gbr_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(test_labels, gbr_pred)))
```

```python
#InicializaÃ§Ã£o do objeto para Grid Search com validaÃ§Ã£o cruzada.
from sklearn.model_selection import GridSearchCV


parameters = {'learning_rate': [0.03],
                  'subsample'    : [0.2],
                  'n_estimators' : [1000],
                  'max_depth'    : [8]
                 }

grid_search = GridSearchCV(grb, parameters, scoring='r2', cv=2, n_jobs=-1)
```

```python
grid_search.fit(train_features, train_labels)
```

```python
    print(" Resultados do Grid Search " )
    print("\n O melhor estimador entre TODOS os parÃ¢metros pesquisados:\n",grid_search.best_estimator_)
    print("\n A melhor pontuaÃ§Ã£o entre TODOS os parÃ¢metros pesquisados:\n",grid_search.best_score_)
    print("\n Os melhores parÃ¢metros entre TODOS os parÃ¢metros pesquisados:\n",grid_search.best_params_)
```
